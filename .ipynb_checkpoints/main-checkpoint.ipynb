```python
import numpy as np
import pandas as pd
```

```python
def make_dict():
    d = {"Startpage":[],
         "Norton":[],
         "start.me":[],
         "Gibiru":[],
         "SearX":[],
         "SAGOOL":[], 
         "yahoo!":[],
         "swisscows":[],
         "Ecosia":[],
         "DuckDuckGo":[],
        }
    return d
```

```python
def input_test_info(d):
    for engine in d:
        arr = np.arange(10)+1
        np.random.shuffle(arr)
        rank = list(arr)
        d[engine] = rank
        print(engine,":",rank)
        print()
    return d

def input_info(d):
    for engine in d:
        print(engine)
        for i in range(10):
            URL = input(str(i)+":")
            d[engine].append(URL)
        print()
    return d

def scoring(d):
    df = pd.DataFrame(np.zeros((len(d), len(d))),
                      index=list(d.keys()),
                      columns=list(d.keys()),
                     )
    M = sum([i**2 for i in range(1,11)])
    for engine_1 in d:
        for engine_2 in d:
            score = 0
            for i in range(len(d)):
                for j in range(len(d)):
                    score += (10-i)*(10-j)*(d[engine_1][i] == d[engine_2][j])
            df.loc[engine_1, engine_2] = score/M
            
    return df
```

```python
d = make_dict()
input_test_info(d)
df = scoring(d)
df
```

```python
d = make_dict()
input_info(d)
df = scoring(d)
df
```
